```python
item_factors_df = pd.read_excel('https://github.com/albanda/CE888/blob/master/lab4-recommender/movies_latent_factors.xlsx?raw=true')
item_factors_df0 = item_factors_df


#import Excel file
user_prefs_df = pd.read_excel('https://github.com/albanda/CE888/blob/master/lab4-recommender/movies_latent_factors.xlsx?raw=true', sheet_name='Users')
user_prefs_df0 = user_prefs_df

user_id = 4768
item_id = 809

user_prefs_df = user_prefs_df.loc[user_prefs_df['User']==user_id]
user_prefs_df = user_prefs_df.drop(['User'], axis=1)
user_prefs = np.array(np.array(user_prefs_df)[0])
print(user_prefs)


item_factors_df = item_factors_df.loc[item_factors_df['Movie ID']==item_id]
item_factors_df = item_factors_df.drop(['Movie ID', 'Title'], axis=1)
item_factors = np.array(np.array(item_factors_df)[0])
print(item_factors)

rating = user_prefs.dot(item_factors)
print(rating)
```

```python
item_factors_df0 = item_factors_df0.drop(['Movie ID', 'Title'], axis=1)
item_factors = np.array(item_factors_df0)
print(item_factors.shape)

user_prefs = user_prefs_df0.loc[user_prefs_df0['User']==4768]

user_prefs = user_prefs.drop(['User'], axis=1)

user_prefs = np.array(user_prefs)[0]

print('User prefs: ', user_prefs)

rating_dict = {}

for i in range(len(item_factors)):#for all movies
  rating = user_prefs.dot(item_factors[i])
  rating_dict[i+1] = rating

sorted_dict = {k: v for k, v in sorted(rating_dict.items(), key=lambda item: item[1])}
print(rating_dict)
print(sorted_dict)
```
